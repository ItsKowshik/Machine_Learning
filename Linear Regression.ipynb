```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Sample values for regression
np.random.seed(100)
x = np.linspace(1,50)
y = 2*x +np.random.normal(0,5,50)
plt.title("Sample scatter for regression")
plt.scatter(x,y)
```

```python
#Data preprocessing step
w = np.zeros((1,1))
b = 0
lr = 0.001
epoch = 100
x = x.reshape((len(x),1))
y = y.reshape((len(y),1))
losses = []
print(x.shape)
print(y.shape)
```

```python
for i in range(epoch):
    y_hat = np.dot(x,w) + b
    loss = np.mean((y_hat-y)**2)
    losses.append(loss)
    dw = np.dot(x.T,y_hat-y)*(1/50)
    db = np.sum(y_hat-y)*(1/50)
    w -= lr*dw
    b -= lr*db

print(losses)
print(w,b)
```

```python
plt.scatter(x,y)
plt.plot(x,np.dot(x,w)+b,'r')
plt.title("Simple Linear Regression")
plt.show()
```

```python
plt.plot([i for i in range(epoch)],losses)
plt.title("Losses for epoch")
plt.show()
```
